```python
import pandas as pd
```

```python
csv_file = "MoviesOnStreamingPlatforms_updated.csv"
streaming_data=pd.read_csv(csv_file)
csv2_file = "TMDb_updated.csv"
overview_data=pd.read_csv(csv2_file)
```

```python
streaming_data.head()
```

```python
streaming_data=streaming_data.drop(columns=['Unnamed: 0', 'Type','Language','Country'])
```

```python
streaming_data["Netflix"].replace({1: True, 0: False}, inplace=True)
streaming_data["Hulu"].replace({1: True, 0: False}, inplace=True)
streaming_data["Prime Video"].replace({1: True, 0: False}, inplace=True)
streaming_data["Disney+"].replace({1: True, 0: False}, inplace=True)
```

```python
streaming_data['Rotten Tomatoes']=streaming_data['Rotten Tomatoes'].str.replace(r'%', r'.0').astype('float')
streaming_data["Title"] = streaming_data["Title"].str.lower()
streaming_data["Title"] = streaming_data["Title"].str.replace(',',' ',regex=True).str.replace(':',' ',regex=True).str.replace('&','and',regex=True)
```

```python
streaming_data=streaming_data.fillna(0)
```
